```python
import pandas as pd
a=pd.read_csv("/content/Salary_Data (1).csv")
print(a)
```

```python
y=a['Salary']
print(y)
```

```python
x=a.drop("Salary" ,axis=1)
print(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.title('SALARY VS YEAR OF EXPERIENCE')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
w0=1
w1=1
yp=w0+(w1*x)
yp=yp.values
print(y,yp)
```

```python
error=[]
for i in range(len(y)):
  error.append((y[i]-yp[i])**2)
import numpy as np
print(np.array(error).mean())
```

```python
e_update=w0+0.01*np.mean(error)
print(e_update)
e_update1=w1+0.01*np.mean(error)
print(e_update1)
```

```python
w0=e_update
w1=e_update1
print(w0)
print(w1)
```

```python
yp=w0+(w1*x)
yp=yp.values
print(y,yp)
```

```python
MAE=np.abs(error).mean()
print(MAE)
```
