```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
housing=pd.read_csv("/content/housing.csv (2).zip")
housing.head()
```

```python
container = ['latitude' , 'total_rooms' , 'housing_median_age' , 'total_bedrooms']
result = housing[container]

mle_params = {}


for i in container:
    mean = np.mean(housing[i])
    variance = np.var(housing[i])
    mle_params = {'mean' : mean , 'variance' : variance}
```

```python
mean=np.mean(housing['latitude'])
print(mean)
variance=np.var(housing['latitude'])
print(variance)

n=len(housing['latitude'])
print(n)
a=-n/2*(math.log(2*3.14))
b=-n/2*(math.log(variance))
c=-1/2*variance

p=50
d=(p-mean)**2
MLE_p=a*b*c*d
print(MLE_p)

q=75
d=(q-mean)**2
MLE_q=a*b*c*d
print(MLE_q)

z=80
d=(z-mean)**2
MLE_z=a*b*c*d
print(MLE_z)
```
